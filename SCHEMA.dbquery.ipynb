```sql
CREATE CATALOG IF NOT EXISTS retail_sales;
CREATE SCHEMA IF NOT EXISTS retail_sales.bronze;
CREATE SCHEMA IF NOT EXISTS retail_sales.silver;
CREATE SCHEMA IF NOT EXISTS retail_sales.gold;

CREATE OR REPLACE TABLE retail_sales.bronze.raw_sales (
  Transaction_ID INTEGER,
  Date_t DATE,
  Customer_ID STRING,
  Gender STRING,
  Age INTEGER,
  Product_Category STRING,
  Quantity INTEGER,
  Price_per_Unit INTEGER,
  Total_Amount INTEGER
)
```
